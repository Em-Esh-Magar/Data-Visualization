```python
# practice of Single bar plot

import matplotlib.pyplot as plt
import numpy as np


x = ["Python","Java","C/C++","Php"]
y = [88,69,45,50]

plt.xlabel("Programming Language")
plt.ylabel("Popularity")

plt.bar(x,y,width=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Python","Java","C/C++","Php"]
y = [88,69,45,50]
z = [56,33,77,69]

width = 0.2

p1 = np.arange(len(x))
p2 = [j+width for j in p1]

plt.xlabel("Programming Language")
plt.ylabel("Popularity")

plt.bar(p1,y,width=0.2)
plt.bar(p2,z,width=0.2)

plt.xticks(p1+width/2,x)

plt.show()
```
