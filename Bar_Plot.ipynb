```python
# practice of Single bar plot

import matplotlib.pyplot as plt
import numpy as np


x = ["Python","Java","C/C++","Php"]
y = [88,69,45,50]

plt.xlabel("Programming Language")
plt.ylabel("Popularity")

plt.bar(x,y,width=0.2)
plt.show()
```
