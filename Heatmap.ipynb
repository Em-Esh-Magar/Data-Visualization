```python
# Temperature of a city
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv('/content/city_temperatures.csv')
data = data.set_index('City')


plt.figure(figsize=(14,8))
sns.heatmap(data,annot=True, cmap='coolwarm')
plt.xlabel('Months')
plt.ylabel('City')
plt.title('Temperature of City')
plt.show()
```
